# **1. Problema de ordenamiento**

El problema de ordenamiento cosiste en una entrada de n numeros desordenados donde la salida de los numeros tiene que estar ordenado de menor a mayor o al rev√©s.

Ejemplo:
```
Entrada:  [5, 2, 4, 6, 1, 3] 
Salida:  [1, 2, 3, 4, 5, 6]
```
image-230-750x298.png
Este problema a sido analizado y resuelto por muchos investigadores, por lo que ha surguido una gran variedad de algoritmos que resuelven el mismo problema, han nacido una serie de conceptos como notaci√≥n de O may√∫scula, algoritmos divide y vencer√°s, estructuras de datos, an√°lisis de los casos peor, mejor, y promedio, y l√≠mites inferiores, esto es bueno para saber cual es mejor algoritmo dependiendo de lo que se necesite.

# **2.Descripci√≥n del algoritmo**

Este tipo de ordenamiento(Quick sort), funciona recursivamente, aparentemente trabaja mejor con elementos completamente desordenados que para elementos semi ordenados, siendo este algoritmo opuesto al bublle sort.

Este algoritmo se basa en la t√©cnica dividir y vencer√°s, lo que lo hace m√°s eficiente que otros algoritmos. Para ordenar, se debe escoger un elemento que est√© en la mitad cerca de la mitad, se empieza a ordenar hasta la mitad, as√≠ se podr√° garantizar que los elementos al la izquierda son menores a los elementos a la derecha, de aqu√≠ en adelante se usan los llamados recursivos para lograr el ordenamiento por partici√≥n, logrando que el tama√±o de los arreglos se reduzca a la mitad y as√≠ continuar hasta que retorne el arreglo ordenado.

Quick sort:

QuickSort recursivo cuando divide el arreglo en 2:


## **2.1.Ejemplo paso a paso**

Elegimos el pivot que esta en medio.

`Arreglo = [8, 3, 6, 4 , 2, 5, 7, 1]`

Paso 1

Escogemos un pivote de nuestro arreglo y un √≠ndice izquierdo y derecho.

`[8, 3, 6, 4, 2, 5, 7, 1]`

En este caso el pivote ser√° `p = 4` los √≠ndices ser√°n `left = 8` y `right = 1` 

Paso 2

`[8, 3, 6, 4, 2, 5, 7, 1]`

El pivote se compara con los √≠ndices si `left > p`, `right < p`-> `rigth` y `left` se intercambian de posici√≥n entre s√≠, 

```
      left = 3
      right = 7
      pivote =4
[1, 3, 6, 4, 2, 5, 7, 8]
```

Paso 3
En el caso de que se cumpla `left < p` y `right > p` para cualquiera de las 2 los √≠ndices se mueven

```
      left = 6(cambio)
      right = 5(cambio)
      pivote =4
[1, 3, 6, 4, 2, 5, 7, 8]
```
El paso 2 y 3 son pr√°cticamente iguales, se ir√°n repitiendo hasta que los √≠ndices lleguen al pivote quedando asi:

`[1, 3, 2, 4, 6, 5, 7, 8]`

paso 4

El arreglo se divide en 2 y se repiten todos los pasos para el sub arreglo izquierdo y derecho obteniendo un resultado final.

`[1, 2, 3, 4, 5, 6, 7, 8]`

Como podemos ver el arreglo queda ordenado al seguir esta serie de pasos.

````python
def partition_der(array, inicio, fin,comp,verbose=False):  
    izquierda = inicio
    Derecha = fin - 1
    pivot = array[fin]
    
    if (verbose == True):
        print("Pivote: ",pivot)
    while izquierda < Derecha:
        #print("paso1")
        
        comp[1] = comp[1] + 1
        while izquierda < fin and array[izquierda] < pivot:
            comp[1] = comp[1] + 1
            izquierda = izquierda + 1
            #print("paso2")
            
        while inicio < Derecha and array[Derecha] >= pivot:
            Derecha = Derecha - 1
            comp[1] = comp[1] + 1
            #print("paso3")    

        if izquierda < Derecha:
          array[izquierda], array[Derecha] = array[Derecha], array[izquierda]
          comp[1] = comp[1] + 1
        else:
          comp[1] = comp[1] + 1
                
    if (array[izquierda] > pivot):
      comp[1] = comp[1] + 1
      if (verbose == True):
        print("Se Intercambian:", array[izquierda], ">", pivot)
      array[izquierda], array[fin] = array[fin], array[izquierda]
    else:
      if (verbose == True):
        print("No Intercambian:", array[izquierda], "<", pivot)
      comp[1] = comp[1] + 1
    if (verbose == True):
      print("Despues del partition: ",array)
    return izquierda

def quickSort_der(array, inicio, fin,comp,verbose=False):
    if inicio < fin:
      comp[0] = comp[0] + 1
      pivote = partition_der(array, inicio, fin,comp,verbose)
      quickSort_der(array, inicio, pivote - 1,comp,verbose)
      quickSort_der(array, pivote + 1, fin,comp,verbose)



````

````python
def merge_sort(array, izquierda, derecha,cont,verbose):
    if izquierda >= derecha:
        return
    medio = (izquierda + derecha)//2
    cont[0] = cont[0] + 1
    merge_sort(array, izquierda, medio,cont,verbose)
    merge_sort(array, medio + 1, derecha,cont,verbose)
    merge(array, izquierda, derecha, medio,verbose)


def merge(array, izquierda, derecha, medio,verbose):
    # Haga copias de ambas array que estamos tratando de fusionar

    # El segundo par√°metro no es inclusivo, por lo que tenemos que aumentar en 1
    cop_izq = array[izquierda:medio + 1]
    cop_der = array[medio+1:derecha+1]

    # Valores iniciales de las variables que usamos para mantener
    # seguimiento de d√≥nde estamos en cada array
    cop_izq_ind = 0
    cop_der_ind = 0
    ord_index = izquierda
    # Pasar por ambas copias hasta que nos quedemos sin elementos en una
    while cop_izq_ind < len(cop_izq) and cop_der_ind < len(cop_der):

        # Si nuestra copia izquierda tiene el elemento m√°s peque√±o, col√≥quelo en el orden
        # part y luego avanzar en left_copy (aumentando el puntero)
        if cop_izq[cop_izq_ind] <= cop_der[cop_der_ind]:

            cont[1] = cont[1] + 1
            array[ord_index] = cop_izq[cop_izq_ind]
            cop_izq_ind = cop_izq_ind + 1


        # Opuesto desde arriba
        else:

            cont[1] = cont[1] + 1
            array[ord_index] = cop_der[cop_der_ind]
            cop_der_ind = cop_der_ind + 1


        # Independientemente de d√≥nde obtuvimos nuestro elemento
        # avanzar en la parte ordenada
        ord_index = ord_index + 1

    # Nos quedamos sin elementos en left_copy o right_copy
    # as√≠ que revisaremos los elementos restantes y los agregaremos

    while cop_izq_ind < len(cop_izq):
        cont[1] = cont[1] + 1
        array[ord_index] = cop_izq[cop_izq_ind]
        cop_izq_ind = cop_izq_ind + 1
        ord_index = ord_index + 1

    while cop_der_ind < len(cop_der):
        cont[1] = cont[1] + 1
        array[ord_index] = cop_der[cop_der_ind]
        cop_der_ind = cop_der_ind + 1
        ord_index = ord_index + 1
````
