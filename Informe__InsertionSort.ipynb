## 1. Problema de ordenamiento

El problema de ordenamiento cosiste en una entrada de n numeros desordenados donde la salida de los numeros tiene que estar ordenado de menor a mayor o al rev√©s.

Ejemplo:
```
Entrada:  [5, 2, 4, 6, 1, 3] 
Salida:  [1, 2, 3, 4, 5, 6]
```

## 2.   Insertion sort

El siguiente c√≥digo muestra una implementaci√≥n del algoritmo Insertion sort con complejidad temporal O(n^2)

En el codigo se ingresa un arreglo de tama√±o n y ordena el arreglo de menor a mayor con los mismos datos de entrada.


### 2.1.C√≥digo

````python
import random
from termcolor import colored
import copy

def insertionSort(arr,verbose=False):
  com = 0
  n = len(arr)
  if (verbose == True): print("Input array: ", arr)

  
  for i in range(1, len(arr) ):
      if verbose == True: print("\nPasada",i)
      already_sorted = True
      j = i           

      while arr[j - 1] > arr[j] and j > 0:
          arr[j - 1], arr[j] = arr[j], arr[j - 1]
          j -= 1
          com+= 1
          
          if verbose == True: 
            print("swap %d:" % com)
            print(str(arr[:j])[1:-1],",",colored(f"{str(arr[j:j+2])[1:-1]}","red"),",",str(arr[j+2:])[1:-1])


      if  verbose == True: 
           print("\nAl finalizar pasada:")
           #print("Pasada",i)
           print("[",colored(f"{str(arr[:i])[1:-1]}","blue") ,",", str(arr[i:])[1:-1],"]")
          
  return arr ,com

com = 0
print("Insertion sort")
arr = [5, 2, 4, 6, 1, 3]

print("Entrada: ",arr,"\n")

(arr , com) = insertionSort(arr)

print("Salida: ",arr)

#print("swap: ",com)
````

## **2.2.Descripci√≥n del algoritmo**

El algoritmo recibe una lista(o arreglo) con cierta cantidad de elementos que queremos ordenar. Los numeros se ordenan en la misma lista(o arreglo)

En palabras simple en cada iteracion se verifica que el numero analizado no tenga un numero mayor a su izquierda.



1.   En cada iterci√≥n el algoritmo verifica que ```arr[j]``` sea mayor a ```0``` tambi√©n compara ``` arr[j-1] > arr[j]```.
2.   Si el algoritmo se encuentra ordenado no realiza cambios por que no se cumple la condicion necesaria.
3.   El algoritmo no requiere de una variable auxiliar aun que existen variantes que si la usan.

## **5. Experimentos**

### **5.1. N√∫mero de comparaciones**



````python
import matplotlib.pyplot as plt

x=[n for n in range(5,20)] 
y1=[n*(n-1)/2 for n in range(5,20)] # worst case
y2=[n-1 for n in range(5,20)] # best case
y=[]; 

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  a,counter = insertionSort(a)
  y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["insertion Sort", "theoretical worst case", "theoretical best case"])

plt.xlabel('n')
plt.ylabel('number of operations')
plt.show()
````

El mejor caso, tiene una una menor cantidad de operaci√≥n que el peor caso, en los casos aleatorios se puede apreciar que esta entre el mejor caso y el peor caso y no tiene un n√∫mero de operaciones constante.
